```python
from mnist_dataloader import MnistDataset, ToTensor
from vae_model import VAE

from mnist_dataloader import MnistDataset, ToTensor
from vae_model import VAE

import torch
import torch.optim as optim
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
import matplotlib.pyplot as plt
import numpy as np

from sklearn.manifold import TSNE
from random import random
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

transform = transforms.Compose([
    transforms.ToTensor(), 
    transforms.Lambda(lambda x: x.view(-1))])

test_dataset = datasets.MNIST(
    "~/mnist", train=False, download=True, transform=transform)

test_dataloader = DataLoader(test_dataset, batch_size=1000, shuffle=False, num_workers=0)

model = VAE(in_size=28, z_dim=10, device=device, hidden_size=200)
model.load_state_dict(torch.load("models/vae_model.pkl"))
model.to(device)

model.eval()
```

```python
fig = plt.figure(figsize=(20, 6))

zs = []
for image, t in test_dataloader:
    for i, im in enumerate(image.view(-1, 28, 28).detach().numpy()[:10]):
        ax = fig.add_subplot(3, 10, i+1, xticks=[], yticks=[])
        ax.imshow(im, 'gray')
        
    image = image.to(device)
    z, y = model(image)
    zs.append(z)
    y = y.view(-1, 28, 28)
    for i, im in enumerate(y.cpu().detach().numpy()[:10]):
        ax = fig.add_subplot(3, 10, i+11, xticks=[], yticks=[])
        ax.imshow(im, 'gray')
        
    z1to0 = torch.cat([z[1, :] * (i * 0.1) + z[0, :] * ((10 - i) * 0.1) for i in range(10)]).reshape(10, 10)
    y2 = model.decoder(z1to0).view(-1, 28, 28)
    for i, im in enumerate(y2.cpu().detach().numpy()[:20]):
        ax = fig.add_subplot(3, 10, i+21, xticks=[], yticks=[])
        ax.imshow(im, 'gray')
    break
```

```python
colors = ["red", "green", "blue", "orange", "purple", "brown", "fuchsia", "grey", "olive", "lightblue"]
def visualize_zs(zs, labels):
    plt.figure(figsize=(10,10))
    points = TSNE(n_components=2, random_state=0).fit_transform(zs)
    for p, l in zip(points, labels):
        plt.scatter(p[0], p[1], marker="${}$".format(l), c=colors[l])
    plt.show()

model.eval()
zs = []
for x, t in test_dataloader:
    x = x.to(device)
    t = t.to(device)
    # generate from x
    y, z = model(x)
    z = z.cpu()
    t = t.cpu()
    visualize_zs(z.detach().numpy(), t.cpu().detach().numpy())
    break
```
